```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
url = 'https://iaryanyadav.github.io/project_1/'
HEADERS = {}
response = requests.get(url, headers=HEADERS)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# ---------- HERO SECTION ----------
hero = soup.find("header", class_="hero")
hero_data = {}
if hero:
    hero_data = {
        "tag": hero.find("span", class_="tag").get_text(strip=True) if hero.find("span", class_="tag") else None,
        "heading": hero.find("h1").get_text(" ", strip=True) if hero.find("h1") else None,
        "description": hero.find("p").get_text(" ", strip=True) if hero.find("p") else None,
        "buttons": [btn.get_text(strip=True) for btn in hero.find_all("button")]
    }
```

```python
# ---------- EVENTS SECTION ----------
events_section = soup.find("section", class_="events")
events_data = []
if events_section:
    cards = events_section.find_all("div", class_="card")
    for card in cards:
        event_tag = card.find("span", class_="event-tag")
        cat_tag = card.find("span", class_="cat-tag")
        brand = card.find("h3")
        desc = card.find("p")
        button = card.find("button", class_="card-btn")

        events_data.append({
            "event_tag": event_tag.get_text(strip=True) if event_tag else None,
            "category": cat_tag.get_text(strip=True) if cat_tag else None,
            "brand": brand.get_text(strip=True) if brand else None,
            "description": desc.get_text(strip=True) if desc else None,
            "button": button.get_text(strip=True) if button else None
        })
```

```python
# ---------- WHY SECTION ----------
why_section = soup.find("section", class_="why")
why_data = {}
if why_section:
    heading = why_section.find("h2").get_text(strip=True) if why_section.find("h2") else None
    description = why_section.find("p").get_text(strip=True) if why_section.find("p") else None

    # Features
    features = []
    feature_boxes = why_section.find_all("div", class_="feature-box")
    for box in feature_boxes:
        icon = box.get_text(strip=True)[0]  # First character = emoji
        title = box.find("h3").get_text(strip=True) if box.find("h3") else None
        desc = box.find("p").get_text(strip=True) if box.find("p") else None
        features.append({
            "icon": icon,
            "title": title,
            "description": desc
        })

    # CTA
    cta_data = {}
    cta = why_section.find("div", class_="cta")
    if cta:
        cta_data = {
            "title": cta.find("h3").get_text(strip=True) if cta.find("h3") else None,
            "description": cta.find("p").get_text(strip=True) if cta.find("p") else None,
            "strong_text": cta.find("strong").get_text(strip=True) if cta.find("strong") else None
        }

    why_data = {
        "heading": heading,
        "description": description,
        "features": features,
        "cta": cta_data
    }
```

```python
# ---------- MERGE ALL DATA ----------
final_data = {
    "hero": hero_data,
    "events": events_data,
    "why": why_data
}
print(final_data)
```

```python
# ---------- SAVE AS JSON ----------
with open("website_data.json", "w", encoding="utf-8") as f:
    json.dump(final_data, f, ensure_ascii=False, indent=4)

print("Data saved to website_data.json")
```
