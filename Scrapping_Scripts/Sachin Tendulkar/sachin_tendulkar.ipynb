```python
import json

# Load your HTML file
with open("sachin_tendulkar.html", "r", encoding="utf-8") as file:
    html_content = file.read()

# Parse HTML
soup = BeautifulSoup(html_content, "html.parser")

# Extract data from 'more' divs
info = {}
for div in soup.find_all("div", class_="more"):
    text = div.get_text(strip=True)
    if ":" in text:
        key, value = text.split(":", 1)
        info[key.strip()] = value.strip()

# Extract data from the table
table = soup.find("table", {"id": "tablepress-7091"})

table_data = []
current_heading = None
content = {}

for row in table.find_all("tr"):
    cells = row.find_all("td")

    # If it's a heading row (colspan="2")
    if len(cells) == 1 and cells[0].get("colspan") == "2":
        # Save previous heading block if exists
        if current_heading and content:
            table_data.append({"heading": current_heading, "content": content})
            content = {}
        current_heading = cells[0].get_text(strip=True)
    elif len(cells) == 2:
        # Normal row: first cell is key, second cell is value
        key = cells[0].get_text(strip=True)
        value = cells[1].get_text(strip=True)
        content[key] = value

# Add the last block after loop ends
if current_heading and content:
    table_data.append({"heading": current_heading, "content": content})
```

```python
# Combine extracted data
data = {
    "name": "Sachin Tendulkar",
    "Personal_info": info,
    "Personal_stats": table_data
}

# Print data as formatted JSON
print(json.dumps(data, indent=4))
```

```python
# Save the data to a JSON file
with open("sachin_tendulkar_data.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4)

print("Data saved to sachin_tendulkar_data.json")
```
