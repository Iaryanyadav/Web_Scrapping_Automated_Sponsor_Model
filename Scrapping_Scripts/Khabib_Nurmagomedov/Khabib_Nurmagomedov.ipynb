```python
# Load your HTML file
with open("Khabib_Nurmagomedov.html", "r", encoding="utf-8") as file:
    html_content = file.read()

soup = BeautifulSoup(html_content, "html.parser")

# Extract tags as a list
tags = [tag.get_text(strip=True) for tag in soup.find_all("p", class_="hero-profile__tag")]

# Extract nickname
nickname = soup.find("p", class_="hero-profile__nickname")
nickname = nickname.get_text(strip=True) if nickname else None

# Extract name
name = soup.find("h1", class_="hero-profile__name")
name = name.get_text(strip=True) if name else None
```

```python
division_block = soup.find("div", class_="hero-profile__division")
division_data = []
# Extract title and body
division_title = division_block.find("p", class_="hero-profile__division-title").get_text(strip=True)
division_body = division_block.find("p", class_="hero-profile__division-body").get_text(strip=True)
division_data = {
    "division": division_title,
    "record": division_body
}
```

```python
stats_block = soup.find("div", class_="hero-profile__stats")
stats = {}

# Loop through each stat div
for stat in stats_block.find_all("div", class_="hero-profile__stat"):
    value = stat.find("p", class_="hero-profile__stat-numb").get_text(strip=True)
    key = stat.find("p", class_="hero-profile__stat-text").get_text(strip=True)
    stats[key] = value
```

```python
img_tag = soup.find("div", class_="hero-profile__image-wrap").find("img")

# Extract alt text and image URL
image_data = {
    "name": img_tag.get("alt"),
    "image_url": img_tag.get("src")
}
```

```python
items = soup.find_all("div", class_="c-carousel__item")

spotlight_data = []
for item in items:
    # Get title
    title_tag = item.find("h3", class_="e-t3")
    title = title_tag.get_text(strip=True) if title_tag else "No title"

    # Get link
    link_tag = item.find("a", href=True)
    link = link_tag["href"] if link_tag else "No link"
    if link != "No link":
        link = "https://www.ufc.com" + link

    # Get image URL (prefer higher resolution if available)
    img_tag = item.find("img")
    image = img_tag["src"] if img_tag else "No image"

    spotlight_data.append({"Title": title, "Link": link, "Image": image})
```

```python
# Store in a dictionary
profile_data = {
    "tags": tags,
    "name": {
        "nickname": nickname,
        "name": name
    },
    "division": division_data,
    "stats": stats,
    "image": image_data,
    "spotlight": {
        "heading": "UFC Spotlight",
        "data": spotlight_data
    }
}

print(json.dumps(profile_data, indent=4))
```
