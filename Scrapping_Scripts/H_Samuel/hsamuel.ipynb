```python
from bs4 import BeautifulSoup
import json

# Load the HTML file
with open("/content/hsamuel.html", "r", encoding="utf-8") as file:  # Replace with your HTML file path
    html_content = file.read()

# Parse with BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Find all swiper-slide divs
slides = soup.find_all("div", class_="swiper-slide")

extracted_data = []

for slide in slides:
    message = slide.find("span", class_="amor-global-banner-message")
    link_tag = slide.find("a", class_="amor-global-banner-link")

    extracted_data.append({
        "message": message.get_text(strip=True) if message else None,
        "link": link_tag["href"] if link_tag else None,
        "link_text": link_tag.get_text(strip=True) if link_tag else None
    })
```
